# 整理爬下來的資料

```python
# 把爬下來的資料放到同個資料夾中
import os
import chardet

# stock_ids = ["1514","2308", "2323", "2609", "2618", "3004", "3017", "3035", "3036", "3231"]

stock_ids = ["2002","2330", "3443"]

# 要處理的資料夾列表
news_folders = ['chinatime', 'ltn']


for stock_id in stock_ids:

    # 遍歷每個資料夾
    for folder in news_folders:
        folder_path = "stock_news/"+stock_id+"/"+stock_id+"_news_"+folder+"/"
        
        content = ''
        files = sorted(os.listdir(folder_path))
        for file_name in files:
            encoding = ''
            with open(folder_path + file_name, 'rb') as f:
                raw_data = f.read()
                encoding = chardet.detect(raw_data)['encoding']
                if encoding != 'utf-8':
                    encoding='Big5'
                # print(chardet.detect(raw_data))
                
            
            with open(folder_path + file_name, 'r', encoding=encoding,errors='ignore') as f:
                content = f.read()
            # print(content)
            
            
            write_folder = "stock_news/"+stock_id+"/"+stock_id+"_news/"

            if not os.path.exists(write_folder):
                os.makedirs(os.path.dirname(write_folder))
                print("建立資料夾" + write_folder)

            with open(write_folder + file_name[6:], "a", encoding="UTF-8") as f:
                f.write(content+"\n")
                print("寫入" + write_folder + file_name[6:])

```

```python
# 整理抓下來的股價資料
import numpy as np
from datetime import datetime
import pandas as pd

stock_ids = ["1101","2211", "2385", "2542", "2880", "2912", "3023", "3264", "5269", "8027"]
# stock_id = "2912"

for stock_id in stock_ids:
    # 讀取 CSV 檔案
    df = pd.read_csv("price_history/"+stock_id+"_price.csv")
    array_np = df.to_numpy()
    
    
    new_array = []
    for item in array_np:
        # print(item)
        
        dateobj = item[0][1:9]
        # print(price)
        i = datetime.strptime(dateobj, '%y/%m/%d').strftime('%Y%m%d')
        value = item[1]
        new_array.append(f"{i},{value}")
        
    print(new_array)
    new_array.reverse()
    # print(new_array_r)
    # print(np.array(new_array_r))
    output_file_path = "price_history/"+stock_id+".csv"
    np.savetxt(output_file_path, new_array, delimiter=',', fmt='%s')
```
