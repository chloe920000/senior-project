```python
#################################################################################
# 更改
# stock_id = '0000'
# glob_url = https://news.tvbs.com.tw/news/searchresult/股票中文名稱/news/
 ################################################################################

stock_id = '2731'
glob_url = 'https://news.tvbs.com.tw/news/searchresult/雄獅/news/'
```

```python
import time
import os
from bs4 import BeautifulSoup
import requests
from datetime import datetime

content = ''
date_last = ''

for i in range(20):
    # 請求網頁內容
    url = glob_url + str(i+1)
    response = requests.get(url)
    
    # 解析網頁內容
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # 找到所有的 <div> 元素並選擇 class="summary" 的元素
    elements = soup.find_all('li')
    
    # 提取每個 summary 元素的文字內容
    for e in elements:
        try:
            # 找到時間元素
            time_text = e.find('div', class_='time').text.strip()
            time_datetime = datetime.strptime(time_text, '%Y/%m/%d %H:%M')
            formatted_date = time_datetime.strftime('%Y%m%d')
            
            if date_last != formatted_date:
                content = ''
            date_last = formatted_date
                
            content = content + e.find('div', class_='summary').text.strip()
            
            folder_path = "stock_news/"+stock_id+"/"+stock_id+"_news_tvbs/"
            file_name = stock_id+"n_" + formatted_date
            if not os.path.exists(os.path.dirname(folder_path)):
                print("建立資料夾" + folder_path)
                os.makedirs(os.path.dirname(folder_path))
            
            with open(folder_path + file_name, "w") as f:
                print("寫入", folder_path + file_name)
                f.write(content)
        except:
            pass
    time.sleep(1)
```
